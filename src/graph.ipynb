```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def plot():
    sample_num = 1000
    artifact_type = ["flower", "plume", "sands", "goblet", "circlet"]
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(1, 1, 1)
    bin_step = 2000 
    range_bin_width = range(0, 20000+bin_step, bin_step)

    array = np.zeros((len(artifact_type), sample_num), dtype=int)
    for i, type in enumerate(artifact_type):
        array[i, :] = np.load(os.path.join("..", "data", type+"_"+"sample="+str(sample_num)+".npy"))
    dist_list = list(array)
    # print(dist_list)

    labels = []
    for i, type in enumerate(artifact_type):
        labels.append(type+" (ave. trial="+str(np.mean(array[i]))+")")
    ax.hist(dist_list, bins=range_bin_width, label=labels)
    ax.set_xlabel("trial")
    ax.legend()
    plt.show()


def main():
    sns.set()
    sns.set_style('whitegrid')
    plot()


if __name__ == '__main__':
    main()
```
